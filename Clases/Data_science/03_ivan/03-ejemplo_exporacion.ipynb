## Exploremos los datos solo con visualizacion

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = sns.load_dataset('titanic')
```

```python
datos.columns
```

```python
datos.info()
```

```python
datos.head()
```

```python
sns.countplot(x='survived',data=datos)
```

```python
# cambiamos de estilo y paletta por defecto
sns.set_style('darkgrid')
sns.set_palette("Set1")
```

```python
sns.countplot(x='survived',data=datos)
```

```python
# veamos la distribucion de edad y sexo de los pasajeros

plt.figure(figsize=(12,8))
sns.stripplot(data=datos,x='sex',y='age')
```

```python
# veamos esta distribucion respecto a las clases en el barco

plt.figure(figsize=(12,8))

#sns.stripplot(data=datos,x='sex',y='age',hue='pclass')

sns.stripplot(data=datos,x='sex',y='age',hue='pclass',dodge=True)
```

```python
# esto nos lleva a ver, cuanta gente habia por clase

sns.countplot(data=datos,x='sex',hue='pclass')
```

```python
sns.countplot(data=datos,x='pclass')
```

```python
# veamos cual es la edad promedio de cada clase discriminado por sexo

#sns.barplot(data=datos,x='pclass',y='age')
sns.barplot(data=datos,x='pclass',y='age',hue='sex')
```

```python
# volvamos a los datos de supervivencia, queremos saber mas detalles sobre los sobrevivientes
```

```python
# veamos la edad promedio de sobrevivientes

sns.barplot(data=datos,x='survived',y='age')
```

```python
sns.stripplot(data=datos,x='survived',y='age')
```

```python
# ver la distribucion de supervivientes por sexo y por clase
```

```python
sns.countplot(data=datos,x='survived',hue='sex')
```

```python
# veamos por sobre las edades
```

```python
sns.histplot(data=datos,x='age',hue='survived',multiple='stack')
```
