Los nombres de las variables tienen que seguir las siguientes reglas y recomendaciones:

    1. Los nombres no pueden empezar con un numero, pero si pueden incluir un numero.
    2. No puede haber espacios en el nombre, usar _ en su lugar.
    3. No se puede usar ninguno de estos simbolos :'",<>/?|\()!@#$%^&*~-+
    4. Esta considerado como buena practica (PEP8) que los nombres esten en minuscula.
    5. Evitar usar caracteres individuales, sobre todo la letra 'l', la letra 'O' o la letra 'I'.
    6. Evitar a toda costa usar palabras que tienen un significado especial en Python, por ejemplo 'list' o 'str'.
    7. Nombrar a las variables con criterio, en relacion a lo que representan.

#### Dynamic Typing

Python usa *dynamic typing*, lo que significa que se pueden reasignar variables a diferentes tipos de datos.

### Determinando el tipo de variable con `type()`
Podemos chequear a que tipo de dato pertenece la variable con la que estamos trabajando usando la funcion predeterminada `type()` . 

Tipos de datos comunes:
* **int** (enteros)
* **float** (flotantes)
* **str** (cadenas)
* **list** (listas)
* **tuple** (tuplas)
* **dict** (diccionarios)
* **set** (conjuntos)
* **bool** (booleanos)
